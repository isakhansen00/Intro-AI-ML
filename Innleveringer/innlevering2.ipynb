```python
#kode for Oppgave 1 her, bruk flere celler hvis det trengs
class FibonacciIterator:
    def __init__(self):
        self.fib_one = 0 # Current two consecutive fibonacci numbers
        self.fib_two = 1 
    
    def __next__(self): # Define the next method
        current = self.fib_one
        self.fib_one, self.fib_two = self.fib_two, self.fib_one + self.fib_two       
        return current

    def __iter__(self):
        return self
    
def main():
    iterator = FibonacciIterator()
    # Display all Fibonacci numbers <= 10000
    fib_list = []
    for i in iterator:
        if i <= 1000: 
             fib_list.append(i)
        else: break
    print(fib_list)

main()
```

```python
#kode for Oppgave 2 her, bruk flere celler hvis det trengs

def fib():
    fib_one = 0 # Current two consecutive fibonacci numbers
    fib_two = 1
    while True:
        current = fib_one
        fib_one, fib_two = fib_two, fib_one + fib_two
        yield current # yields the current Fibonacci number

def main():
    iterator = fib()
    # Displays all Fibonacci numbers <= 1000
    fib_list = []
    for i in iterator:
        if i <= 1000: 
             fib_list.append(i)
        else: break
    print(fib_list)

main()
```
