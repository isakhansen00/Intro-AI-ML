```python
class MyStack(MyLinkedList):
    def __init__(self, data):
        super().__init__(data)

    def push(self, data):
        self.append(data)

    def top(self):
        return self.tail

    def peek(self):
        return self.top()

    def pop(self):
        temp = self.tail
        self.tail = temp.prev
        self.tail.next = None
        return temp.get_content()

def main():
    stack = MyStack('e')

    # append() function to push
    # element in the stack

    stack.push('a')
    stack.push('b')
    stack.push('c')

    print('Initial stack')
    for elem in stack:
        print(elem)

    # pop() function to pop
    # element from stack in

    # LIFO order
    print('\nElements popped from stack:')
    print(stack.pop())
    print(stack.pop())
    print(stack.pop())

    print('\nStack after elements are popped:')
    for elem in stack:
        print(elem)

    # uncommenting print(stack.pop()) should cause an IndexError as the stack is now empty

main()
```

```python
class Stack():
    def __init__(self):
        self.data = []
    
    def get_content(self):
        return self.data

    def push(self, data):
        self.data.append(data)

    def top(self):
        return self.data[:-1]

    def peek(self):
        return self.top()

    def pop(self):
        data = self.data.pop()
        return data

    def __str__(self) -> str:
        return f"{str(self.get_content())}"

def main():
    stack = Stack()
    # append() function to push
    # element in the stack

    stack.push('a')
    stack.push('b')
    stack.push('c')

    print('Initial stack')
    print(str(stack))

    # pop() function to pop
    # element from stack in

    # LIFO order
    print('\nElements popped from stack:')
    print(stack.pop())
    print(stack.pop())
    print(stack.pop())

    print('\nStack after elements are popped:')
    print(stack)

main()
```
